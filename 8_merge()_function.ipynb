# Merge Function 

## Q - Fetch name of batsman witch is score highest is season

```python
import numpy as np
import pandas as pd 
```

#### it help to merge two dataframe with at least one similer column

```python
match = pd.read_csv('./DataSets/ipl_match.csv')
delivery = pd.read_csv('./DataSets/deliveries.csv')
```

```python
new = delivery.merge(match, left_on = 'match_id', right_on = 'id')
```

```python
new.head(2)
```

```python
new.groupby(['season', 'batsman'])['batsman_runs'].sum().sort_values(ascending = False).reset_index().drop_duplicates('season').sort_values('season')
```
